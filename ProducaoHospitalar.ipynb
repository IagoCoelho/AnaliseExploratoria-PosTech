```python
#StringIO(entrada)input/(saida)output
from io import StringIO

dados_da_populacao_io = StringIO(dados_da_populacao)

populacao = pd.read_csv(dados_da_populacao_io, sep="\t")
populacao = populacao.dropna()
populacao.head()
```

Limpar as colunas, deixar a penas UF do brasil e população

```python
#Pegando as colunas da tabela 
populacao.columns = ["posicao", "uf", "populacao", "porcentagem", "pais_comparevel"]
#                          fazendo replace dentro da series e transformando em int 
populacao["populacao"] = populacao["populacao"].str.replace(" ", "").astype(int)
populacao = populacao[["uf", "populacao"]]
populacao.head()
```

```python
display(gastos_mais_recente.head())
display(populacao.head())
```

```python
# fazendo set no index
populacao = populacao.set_index("uf")
populacao.head()
```

```python
# tirando os espaços em branco  strip --> tira os espaços em branco tanto da direita quanto da esquerda
populacao.index = populacao.index.str.strip()
populacao.head()
```

```python
#ignorando os 3 primeiros caracteres do indice: [3:] --> ignorando os 3 primeiros caracteres
gastos_mais_recente.index = gastos_mais_recente.index.str[3:]
gastos_mais_recente.head()
```

```python
# esta entrando na coluna estado e substituindo todos os estados com nome duplicado por so um

for estado in gastos_mais_recente.index:
    populacao.index = populacao.index.str.replace(f"{estado} {estado}", estado)
populacao.index
```

```python
gastos_populacao = populacao.join(gastos_mais_recente)
gastos_populacao
```

```python
ultima_coluna = gastos_populacao.columns[-1]
gastos_populacao["gastos"] = gastos_populacao[ultima_coluna] * 1_000_000
gastos_populacao.plot()
```

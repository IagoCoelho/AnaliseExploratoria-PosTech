```python
import pandas as pd
```

```python
dados = pd.read_csv("ProducaoHospitalar.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, thousands=".", decimal=",", na_values=['NA', 'N/A', '-'])
```

```python
#mostra as 5 primeiras linhas do arquivo
dados.head()
```

```python
#mostra as 5 ultimas linhas do arquivo
dados.tail()
```

```python
#faz a media de cada coluna
media = dados.mean(numeric_only=True)
media
```

```python
#traz as informações que contem no arquivo
dados.info()
```

```python
#ira fazer a formatação dos numerais
pd.options.display.float_format = "{:.2f}".format
```

```python
media
```

```python
dados["2008/Ago"]
```

```python
#ira tirar a media apenas da coluna de Agosto de 2008
dados["2008/Ago"].mean()
```

```python
#ira fazer o plot(criar grafico)
dados.plot(x="Unidade da Federação",y="2008/Ago")
```

```python
#kind sera o tipo de grafico que sera usado
dados.plot(x="Unidade da Federação", y="2008/Ago", kind = "bar")
```

```python
#figsize é o tamanho que ficara ografico
dados.plot(x="Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(9,6))
```

```python
import matplotlib.ticker as ticker

axis = dados.plot(x="Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
```

```python
import matplotlib.pyplot as plt

import matplotlib.ticker as ticker

axis = dados.plot(x="Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade da federação")
plt.show()
```
